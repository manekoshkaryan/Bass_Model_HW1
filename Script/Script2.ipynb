```python
df_cleaned['Year'] = df_cleaned['Year'] - df_cleaned['Year'].min()
df_cleaned.rename(columns={'LCD TV': 'Count'}, inplace=True)
print(df_cleaned)
```

# Point 4

```python
x_data = df_cleaned["Year"].values
y_data = df_cleaned["Count"].values


try:
    params, _ = curve_fit(bass_model, x_data, y_data, p0=[0.02, 0.28, max(y_data)])
    p_est, q_est, M_est = params

    print(f"Estimated Parameters:")
    print(f"p (innovation)  = {p_est:.4f}")
    print(f"q (imitation)   = {q_est:.4f}")
    print(f"M (market size) = {M_est:.2f}")
except Exception as e:
    print(f"Error in curve fitting: {e}")
```

# Point 5

```python

x_fit = np.linspace(x_data.min(), x_data.max(), 100)
y_fit = bass_model(x_fit, p_est, q_est, M_est)


x_future = np.linspace(x_data.max() + 1, x_data.max() + 10, 10)
y_future = bass_model(x_future, p_est, q_est, M_est)

x_full = np.concatenate([x_fit, x_future])
y_full = np.concatenate([y_fit, y_future])

plt.figure(figsize=(8,5))
plt.scatter(x_data, y_data, label="Observed Data", color="blue")
plt.plot(x_fit, y_fit, label="Bass Model (Historical Fit)", color="red")
plt.plot(x_future, y_future, label="Bass Model (Future Prediction)", color="green", linestyle="--")
plt.xlabel("Time (Years since first observation)")
plt.ylabel("New Adopters per Year")
plt.title("Bass Model: Historical Fit + Future Diffusion Prediction")
plt.legend()
#plt.savefig("Bass_model_vizualisation.png", dpi=300, bbox_inches="tight")
plt.show()


cumulative_adopters = np.cumsum(y_full)
plt.figure(figsize=(8,5))
plt.plot(x_full, cumulative_adopters, label="Cumulative Adopters", color="purple")
plt.axhline(M_est, label="Market Potential (M)", color="black", linestyle=":")
plt.xlabel("Time")
plt.ylabel("Cumulative Adopters")
plt.title("Cumulative Adoption Curve")
plt.legend()
#plt.savefig("Cumulative_adoption_curve.png", dpi=300, bbox_inches="tight")
plt.show()
```

# Point 7

```python
time_horizon = np.arange(0, 11) 
new_adopters = [bass_model(t, p_est, q_est, M_est) for t in time_horizon]
cumulative_adopters = np.cumsum(new_adopters)
forecast_data = []
for i, t in enumerate(time_horizon):
    forecast_data.append({
        "Period (t)": int(t),
        "New Adopters": round(new_adopters[i], 2),
        "Cumulative Adopters": round(cumulative_adopters[i], 2)
    })

df_forecast = pd.DataFrame(forecast_data)
print(df_forecast)
```
